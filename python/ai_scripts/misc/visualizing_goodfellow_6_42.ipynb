```python
import math
import matplotlib.pyplot as plt

def rectifier_complexity(d, l, n):
    """
    Calculates the Big-Oh complexity of a deep rectifier network.
    """
    return (math.comb(n, d) ** (d * (l-1))) * (n ** d)

# Define the range of values for d, l, and n
d_values = range(2, 8)
l_values = range(2, 6)
n_values = range(2, 6)

# Calculate the complexity for each combination of d, l, and n
complexities = []
for d in d_values:
    for l in l_values:
        for n in n_values:
            complexities.append(rectifier_complexity(d, l, n))

# Plot the results
plt.scatter(range(len(complexities)), complexities, alpha=.8)
plt.xlabel('Network configurations')
plt.ylabel('Complexity')
plt.title('Big-Oh complexity of deep rectifier networks')
plt.show()
```

```python
from math import comb
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def complexity(d: int, l: int, n: int) -> int:
    return (comb(n, d))**(d*(l-1)) * n**d

d_values = np.arange(1, 5)
l_values = np.arange(1, 5)
n_values = np.arange(1, 5)

X, Y = np.meshgrid(d_values, l_values)
Z = np.zeros((len(d_values), len(l_values)))

for i in range(len(d_values)):
    for j in range(len(l_values)):
        Z[i][j] = complexity(d_values[i], l_values[j], n_values[0])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

ax.set_xlabel('d values')
ax.set_ylabel('l values')
ax.set_zlabel('Complexity')

plt.show()
```
